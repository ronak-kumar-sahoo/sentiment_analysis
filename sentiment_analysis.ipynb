```python
import pandas as pd
import numpy as np
import nltk
import io
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
nltk.download('punkt')
nltk.download('stopwords')
from transformers import pipeline
from google.colab import files

uploaded = files.upload()  # Upload CSV file
file_name = next(iter(uploaded))
df = pd.read_csv(io.BytesIO(uploaded[file_name]), encoding='latin1')

# Drop rows with missing values in 'text' or 'sentiment'
df.dropna(subset=['text', 'sentiment'], inplace=True)

# Use the transformers pipeline for sentiment analysis
classifier = pipeline('sentiment-analysis')

# Apply the pipeline to the entire 'text' column
# Since the pipeline can be slow on large datasets, we'll process in batches
batch_size = 100
sentiment_results = []
for i in range(0, len(df), batch_size):
    batch_texts = list(df['text'][i:i+batch_size].astype(str))
    batch_sentiments = classifier(batch_texts)
    if batch_sentiments:  # Check if the batch_sentiments list is not empty
        sentiment_results.extend(batch_sentiments)
    else:
        # If the batch is empty or pipeline returns no result, append placeholders
        sentiment_results.extend([{'label': None, 'score': None}] * len(batch_texts))


# Extract labels and scores
predicted_sentiments = [result['label'] for result in sentiment_results]
sentiment_scores = [result['score'] for result in sentiment_results]

# Add the results to the DataFrame
df['predicted_sentiment_label'] = predicted_sentiments
df['predicted_sentiment_score'] = sentiment_scores

# Display the first few rows with the new columns
display(df.head())
```

```python
# Sentiment Analysis using RNN on IMDB Dataset
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
# -----------------------------
# Step 1: Load and Prepare Data
# -----------------------------
num_words = 10000   # Use top 10,000 most frequent words
maxlen = 200
# Each review truncated/padded to 200 words
print("Loading IMDB dataset...")
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=num_words)
print(f"Training samples: {len(x_train)}, Testing samples: {len(x_test)}")
# Pad sequences to ensure equal length
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)
# -----------------------------
# Step 2: Build RNN Model
# -----------------------------
model = Sequential([
Embedding(input_dim=num_words, output_dim=32, input_length=maxlen),
SimpleRNN(32, dropout=0.2, recurrent_dropout=0.2),
    Dense(1, activation='sigmoid')
])

model.summary()

# -----------------------------
# Step 3: Compile the Model
# -----------------------------
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# -----------------------------
# Step 4: Train the Model
# -----------------------------
print("\nTraining the RNN model...")
history = model.fit(
    x_train, y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)

# -----------------------------
# Step 5: Evaluate the Model
# -----------------------------
loss, acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\nTest Accuracy: {acc * 100:.2f}%")

# -----------------------------
# Step 6: Predict Sample Reviews
# -----------------------------
word_index = imdb.get_word_index()
reverse_index = {v: k for k, v in word_index.items()}
def decode_review(encoded_review):
    return ' '.join([reverse_index.get(i - 3, '?') for i in encoded_review])

# Select a single sample review from the test set
sample_review = x_test[0]
decoded_review = decode_review(sample_review)

# Predict the sentiment for the single sample review
prediction = model.predict(np.array([sample_review]))[0][0]

print("\nSample Review:")
print(decoded_review)
print(f"\nPredicted Sentiment: {'Positive' if prediction > 0.5 else 'Negative'} ({prediction:.3f})")
```
